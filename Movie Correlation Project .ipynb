```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) #Adjusting the condition of the plot

df= pd.read_csv(r'C:\Users\danny\Downloads\movies.csv')
```

```python
#Data visualized
df.head()
```

```python
#Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Data types
df.dtypes
```

```python
#Change data type and handles missing values
df['budget'] = df['budget'].fillna(0).astype(int)
df['gross'] = df['gross'].fillna(0).astype(int)
```

```python
df.head()
```

```python
#Sort Data by Gross
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)

df.head()
```

```python
# Scatter plot with Budget vs Gross

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color' : 'red'}, line_kws = {'color' : 'blue'})
```

```python
# Create a temporary DataFrame with numeric data only
df_temp = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = df_temp.corr(method = 'pearson') #pearson, spearman

# Display the correlation matrix
print(correlation_matrix)
```

```python
# Heatmap for Correlation
sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at Company
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
# New heatmap with new var
correlation_matrix = df_numerized.corr(method = 'pearson') #pearson, spearman

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
